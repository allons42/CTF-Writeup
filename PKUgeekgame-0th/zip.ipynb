```python
import binascii
def hex2a(s):
    s = s.split(' ')
    ss = ""
    for x in s:
        c = binascii.a2b_hex(x)
        ss += c.decode()
    return ss
```

```python
f = open("quine.zip",'rb')
x = f.read(0x21f)
```

```python
for i in range(0, 0xffffffff):
    si = hex(i)[2:]
    si = (8-len(si))*"0" + si
    hi = [binascii.a2b_hex(si[0:2]), binascii.a2b_hex(si[2:4]), binascii.a2b_hex(si[4:6]), binascii.a2b_hex(si[6:8])]
    hi = hi[::-1]
    for p in pos:
        x = x[0:p] + hi[0] + hi[1] + hi[2] + hi[3] + x[p+4:]
    if zlib.crc32(x) == i:
        print("got it! ",i)
    if (i % 1e6 == 0):
        print(i)
```
