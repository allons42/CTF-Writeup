```python
import pickle
with open("wrong_ans.pkl", "rb") as f:
    wrong_ans, correct_ans = pickle.load(f)

def is_prime(x):
    if (x % 6 != 1) and (x % 6 != 5):
        return False
    for i in range(5, int(x ** 0.5) + 1, 6):
        if (x % i == 0) or (x % (i + 2) == 0):
            return False
    return True

def get_prime(a,b):
    res = []
    for num in range(a, b+1):
        if is_prime(num):
            res.append(num)
            
    for x in correct_ans:
        if x in res:
            return [x]
    
    res = [x for x in res if x not in wrong_ans]
    return res
```

```python
import re
from random import choice

def getAns(Q):
    mem = None
    if "bilibili" in Q:
        ans = "418645518\n"
    elif "gStore" in Q:
        ans = "10.14778/2002974.2002976\n"
    elif "HTTP" in Q:
        ans = "ctf.xn--4gqwbu44czhc7w9a66k.com\n"
    elif "WebP" in Q:
        ans = "65\n"
    elif "Android" in Q:
        ans = "cn.edu.pku.pkurunner\n"
    elif "质数" in Q:
        nums = re.findall(r"\d+",Q)
        primes = get_prime(int(nums[1]), int(nums[2]))
        mem = choice(primes)
        ans = f"{mem}\n"
    elif "关卡" in Q:
        level = int(re.findall(r"\d+",Q)[1])
        ans = f"{dic[level]}\n"
    elif "邮编" in Q:
        ans = "99645\n"
    else:
        ans = "what?\n"
    return ans,mem
```
