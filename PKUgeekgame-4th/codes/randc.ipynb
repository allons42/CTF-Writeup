```python
import numpy as np
from scipy.optimize import nnls

def calc(n):
    mat = np.zeros((5*n, n))
    b = np.zeros(5*n)

    for i in range(5*n):
        # i: 0 - (n-1)，第i个方程

        mat[i, i%n] = 1
        mat[i, (i+28)%n] = 1
        mat[i, (i+31)%n] = -1

        b[i] = interval[i]
        
    x = nnls(mat, b)
    return x
```

```python
for i in range(6,80):
    print(i, calc(i))
```

```python
x = calc(68)

for i in x[0]:
    t = round(i)
    print(chr(t+1), end="")
print()
for i in x[0]:
    t = round(i)
    print(chr(t), end="")
print()
for i in x[0]:
    t = round(i)
    print(chr(t-1), end="")
```
